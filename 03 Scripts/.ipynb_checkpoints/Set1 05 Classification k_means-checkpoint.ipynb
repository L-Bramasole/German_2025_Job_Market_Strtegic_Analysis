# Imporing Libraties and data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.cluster import KMeans # Here is where you import the k-means algorithm from scikit-learn.
import pylab as pl # PyLab is a convenience module that bulk imports matplotlib.
```

```python
# This option ensures the graphs you create are displayed in your notebook without the need to "call" them specifically.

%matplotlib inline
```

```python
Jobmarket01 = pd.read_excel(r'C:\Users\analy\Documents\German 2025 Job Market Analysis\02 Data\Manipulated\Job Market Germany data set.xlsx')
```

```python
Jobmarket01.columns
```

```python
# creating subset dataframe that include only the numeric columns and the encoded columns of the categorial ones
Jobmarket01_encoded = Jobmarket01[
    [
        # numeric columns you specified
        'salary_median_eur',
        'number_of_openings',
        'demand_index',
        'competitiveness_score',

        # encoded categorical columns
        'job_title_encoded',
        'industry_encoded',
        'city_encoded',
        'state_encoded',
        'education_requirement_encoded',
        'job_type_encoded',
        'skills_required_encoded',
        'hiring_trend_encoded',
        'season_encoded',
        'experience_level_encoded',
        'remote_availability_encoded'
    ]
]
```

```python
Jobmarket01_encoded.head()
```

```python
# save the encoded dataframe in case i will use it later
Jobmarket01.to_excel(r'C:\Users\analy\Documents\German 2025 Job Market Analysis\02 Data\Manipulated\Job Market Germany dataset_encoded.xlsx')
```

# 2. The elbow technique

```python
cluster_r= range(1,10)
```

```python
kmeans1= [KMeans(n_clusters=i) for i in cluster_r] # calculating kmean to all clusters from 1 to 10
```

```python
# Create a score that represents a rate of variation for the given cluster option.
score1 = [kmeans1[i].fit(Jobmarket01_encoded).score(Jobmarket01_encoded) for i in range(len(kmeans1))]

score1
```

```python
# Plot the elbow curve using PyLab.

pl.plot(cluster_r,score1)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

There is a slow down in the score after cluster number 3. Therefore, 3 clusters will be taken for prediction

# 3. k-means2 clustering

```python
# Create the k-means object.

kmeans2 = KMeans(n_clusters = 3) 
```

```python
kmeans2.fit(Jobmarket01_encoded)
```

```python
Jobmarket01_encoded['clusters']= kmeans2.fit_predict(Jobmarket01_encoded)
```

```python
Jobmarket01_encoded.head()
```

```python
Jobmarket01_encoded['clusters'].value_counts()
```

# 4. rescaling  the dataframe variables with standardscaling

```python
# rescaling all the dataframe variables using standard scaling that to make al the features contribute equally
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
Jobmarket01_scaled = scaler.fit_transform(Jobmarket01_encoded)
```

```python
# the rescaling created and numPy array instead of the panda dataframe.
# here i am converting the array back to dataframe to continue my analysis

# Convert scaled array back to DataFrame with original column names
Jobmarket01_scaled = pd.DataFrame(Jobmarket01_scaled, columns=Jobmarket01_encoded.columns)
```

```python
Jobmarket01_scaled.head()
```

# 5. The elbow technique for scaled dataframe

```python
# repeating the elbow technique to decide on the number of clusters based on the rescaled dataframe
cluster_r2= range(1,10)
```

```python
kmeans3= [KMeans(n_clusters=i) for i in cluster_r2]
```

```python
# Create a score that represents a rate of variation for the given cluster option.
score2 = [kmeans3[i].fit(Jobmarket01_scaled).score(Jobmarket01_scaled) for i in range(len(kmeans3))]

score2
```

```python
# Plot the elbow curve using PyLab.

pl.plot(cluster_r2,score2)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

There is a slow down in the score after cluster number 3 but the line is still going up and not flattened after that. Therefore, 3 clusters will be taken for prediction

the heatmap is showing how each clusterâ€™s average feature values compare to the overall dataset mean. using StandardScaler (mean = 0, standard deviation = 1), means:

A value of 0 = the feature is exactly at the datasetâ€™s average.

Positive values (e.g. +1) = the cluster centroid is 1 standard deviation above the mean for that feature.

Negative values (e.g. -2) = the cluster centroid is 2 standard deviations below the mean for that feature.

Dark red (positive) â†’ that cluster has higher-than-average values for that feature.

Dark blue (negative) â†’ that cluster has lower-than-average values for that feature.

Near 0 â†’ that cluster is close to the dataset average for that feature.

Example:

If salary_median_eur shows +1 for cluster 2 â†’ jobs in that cluster have salaries about 1 standard deviations above the dataset average.

If competitiveness_score shows -1.0 for cluster 0 â†’ jobs in that cluster are less competitive than average.


### German Job Market 2025 Classification â€“ Heatmap Analysis 

In analyzing the German job market of 2025, a heatmap was created using Kâ€‘Means clustering, which grouped the data into three distinct clusters. Each cluster highlights different characteristics of the labor market: 

Cluster 0  

This cluster reflects a segment with low hiring trends, low demand index, low competitiveness scores, and relatively fewer job openings. It represents areas of the market where opportunities are limited and competition is weak. 

Cluster 1  

Here we see jobs with a relatively low median salary and lower experience requirements, but at the same time, a high number of job openings, strong hiring trends, and a high demand index. This suggests that while compensation is modest, opportunities are abundant and demand is strong. 

Cluster 2  

This cluster is characterized by very high median salaries, high experience levels, and high competitiveness scores, combined with strong hiring trends and high demand index. However, it shows a relatively lower number of job openings, indicating that these positions are lucrative but scarce and highly competitive. 

Overall Insight 

The analysis reveals that job type, job title, industry, city, state, education level, and remote availability do not significantly influence the number of job openings across these clusters. Instead, the clusters are primarily differentiated by salary levels, experience requirements, competitiveness, and demand dynamics.
